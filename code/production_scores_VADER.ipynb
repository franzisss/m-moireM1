```python
nltk.download('punkt')

df=pd.read_csv('/content/marc_gotti.csv') #on met le chemin vers le fichier

text=' '.join(df['chunk_text'])

# Tokeniser le texte en segments (phrases)
chunks = chunk_text(text)

# Initialiser l'analyseur de sentiments VADER
analyzer = SentimentIntensityAnalyzer()

# Initialiser le traducteur Google
translator = Translator()

# Traduire chaque segment du français vers l'anglais
translated_chunks = [translator.translate(chunk, src='fr', dest='en').text for chunk in chunks]

# Calculer les scores composés pour chaque segment traduit
compound_scores = [analyzer.polarity_scores(chunk)['compound'] for chunk in translated_chunks]

# Fonction pour obtenir l'étiquette de sentiment basée sur le score composé
def get_sentiment_label(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Étiqueter les sentiments
sentiment_labels = [get_sentiment_label(score) for score in compound_scores]

# Créer un DataFrame avec les résultats
df_result = pd.DataFrame({
    'original_chunk': chunks,
    'translated_chunk': translated_chunks,
    'compound_score': compound_scores,
    'sentiment_label': sentiment_labels
})
df_result
```

```python
df_result.to_csv('marc_gotti_trad.csv', index=False)
```
